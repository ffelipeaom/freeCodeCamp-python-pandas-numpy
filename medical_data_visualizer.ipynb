```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/medical_examination.csv")
df.head()
```

# Add 'overweight' column

```python
#calculate their BMI by dividing their weight in kilograms by the square of their height in meters.
#If that value is > 25 then the person is overweight.
#Use the value 0 for NOT overweight and the value 1 for overweight.
#df_overweight = df
#df_bmi = pd.DataFrame(round((df["weight"] / (df["height"]/100)), 2)).rename({0: "bmi"}, axis='columns')
df["bmi"] = round((df["weight"] / (df["height"]/100)), 2)
df.head()
```

```python
#df_overweight["overweight"] = df_bmi.query('bmi >= 25')
df["overweight"] = list(map(int, df["bmi"] >= 25))
df.head()
```

```python
df["overweight"].value_counts()
```

# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.

```python
df[["gluc", "cholesterol"]].describe() #checking min and max values
```
